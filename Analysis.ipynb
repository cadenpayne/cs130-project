```python
all_state = []

for line in open('state_crime.csv'):
    line = line.strip()
    state_data = line.split(",")
    all_state.append(state_data)
```

```python
header = all_state[0]
rows = all_state[1:]

print("\nHeader:")
print(header)

print("\nNumber of data rows:")
print(len(rows))
for row in all_state[1:10]:
    print(row[0])
```

```python
def round_data_frame(d):
    return round(d)

my_data.Rates_Violent_All.apply(round_data_frame)
```

```python
# lambda functions
# lambda d: round(d, 1) : for the above function
my_data.Rates_Violent_All.apply(lambda d: round(d))
```

```python
# Sorting
my_data.sort_values(by='Population', ascending=False).head()
```

```python
import matplotlib.pyplot as plt

my_data.Rates_Property_All.plot(kind='hist')
plt.title('Property Crime Data')
plt.xlabel('Number of Property Crimes')
```

```python
my_data.plot(kind='scatter', x='Year', y='Rates_Property_All', figsize=(12,6), s=1)
plt.title("Title")
plt.xlabel("Year")
plt.ylabel("Property Crime Rates")
```
